```python
#polynomial regression
def plotpoly(model,indevar,devar, name):
    xnew= np.linspace(15,55,100)
    ynew= model(xnew)
    plt.plot(indevar,devar,'.',xnew,ynew,'-')
    plt.title("Polynomial Fit")
    ax=plt.gca()
    ax.set_facecolor((0.898,0.898,0.898))
    fig= plt.gcf()
    plt.xlabel(name)
    plt.ylabel('price of car')
    plt.show()
    plt.close()
```
