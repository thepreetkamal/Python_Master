```python
#plt.subplot()- discussed earlier(matlab approach,not OOP)
#plt.subplots()- OO approach-more control over aesthetics and data
```

```python
fig, ax= plt.subplots(1,2) #1 row,2 columns. The function returns a figure object and a tuple containing axes objects equal to nrows*ncols. Each axes object is accessible by its index
#Brute force approach
ax[0].plot(x,y,"g")
ax[1].plot(x,y,"c")
```

```python
fig, ax= plt.subplots(1,2) #1 row,2 columns. The function returns a figure object and a tuple containing axes objects equal to nrows*ncols. Each axes object is accessible by its index
#for loop approach
for axes in ax:
    axes.plot(x,y)
```

```python
fig, ax= plt.subplots(1,2) 
col=["g","c"]
data=[y,x*y]
#for loop approach
for i, axes in enumerate(ax):
    axes.plot(x,data[i],col[i])
```

```python
fig, ax= plt.subplots(1,2) 
col=["g","c"]
data=[y,x*y]
#for loop approach
for i, axes in enumerate(ax):
    axes.plot(x,data[i],col[i])
fig.tight_layout() #automatically adjusts fig,avoiding overlaps
```

```python
#change fig size n aspect ratio
fig, ax1= plt.subplots(figsize=(8,4),dpi=100)
ax1.plot(x,y,"g",label="y")
ax1.plot(x,y*x,"c",label="y*x")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("Random plot")
#adding legend
ax1.legend()
#to save this fig
fig.savefig("randomfile.png",dpi=100)
```

```python
#customise legend location
fig, ax1= plt.subplots(figsize=(8,4),dpi=100)
ax1.plot(x,y,"g",label="y")
ax1.plot(x,y*x,"c",label="y*x")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("Random plot")
ax1.legend(loc=5) #nos from 0 to 5
fig.savefig("randomfile.png",dpi=100)
```

```python
#advanced matplotlib concepts
fig, ax= plt.subplots(1,2,figsize=(10,4))
ax[0].plot(x,y,x,x*y)
ax[1].plot(x,np.exp(x))
fig.tight_layout()
```

```python
fig, ax= plt.subplots(1,2,figsize=(10,4))
ax[0].plot(x,y,x,x*y)
# to make y as log scale
ax[1].set_yscale('log')
ax[1].plot(x,np.exp(x))
fig.tight_layout()
```

```python
#to set,reset and resize the ticks
fig, ax= plt.subplots(figsize=(10,4))
ax.plot(x,x*y,lw=5,color='g')
ax.set_xticks([1,10,20]) #we want ticks for points (1,10,20)
fig.tight_layout()
```

```python
#to set,reset and resize the ticks
fig, ax= plt.subplots(figsize=(10,4))
ax.plot(x,x*y,lw=5,color='g')
ax.set_xticks([]) #no ticks
fig.tight_layout()
```

```python
#to set,reset and resize the ticks
fig, ax= plt.subplots(figsize=(10,4))
ax.plot(x,x*y,lw=5,color='g')
ax.set_xticks(x) #we want ticks for all points
fig.tight_layout()
```

```python
#set labels acc to ticks
fig, ax= plt.subplots(figsize=(10,4))
ax.plot(x,x*y,lw=5,color='g')
ax.set_xticks([1,10,20]) #we want ticks for points (1,10,20)
ax.set_xticklabels(['a','b','c'],fontsize=24)
```

```python
from matplotlib import ticker
```

```python
#to print y in scientific notation
fig, ax= plt.subplots(figsize=(10,4))
ax.plot(x,x*y,lw=5,color='g')
ax.set_xticks([1,10,20])
#ax.set_yticks([0,500])
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(formatter)
ax.set_xticklabels(['a','b','c'],fontsize=24)
```
