```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style="darkgrid")
```

```python
tips =sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.tail()
```

```python
tips.info()
```

```python
#how tips and total bill are correlated
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
dir(sns.FacetGrid) # press o to hide output or insert; the end
```

```python
#how smoker/non-smoker affets the total bill vs tips
sns.relplot(x='total_bill',y='tip',data= tips,hue='smoker')#here we have introduced smoker based on categorical values
```

```python
tips['smoker'].value_counts()
```

```python
#relationship between 4 variables
sns.relplot(x='total_bill',y='tip',data= tips,hue='smoker',style='time')
#here we have introduced smoker & time based on categorical values. The colour scheme used is categorical palette
```

```python
#use colour with different intensity
sns.relplot(x='total_bill',y='tip',data= tips,hue='size')
#here we have introduced the references based on numerical values.The colour scheme used is qualitative palette since these are easier to use with continuous numerical data. Here we can change the size, colour,width of the circle etc
```

```python
#change colour of the circle
sns.relplot(x='total_bill',y='tip',data= tips,hue='size',palette='ch:r=-0.5,l=0.95')
```

```python
#change size of the circle
sns.relplot(x='total_bill',y='tip',data= tips,size='size',palette='ch:r=-0.5,l=0.95') #earlier hue='size' that means colour was changing according to the size
```

```python
#increase the size of dataplots.these vsalues are normalised.we will provide custom sizes, sns will normalise them and plot them
sns.relplot(x='total_bill',y='tip',data= tips,size='size',sizes=(50,200),palette='ch:r=-0.5,l=0.95')
```

```python
fmri= sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
fmri.info()
```

```python
#plotting relplot
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri)
```

```python
#the above line is computed with 95 percent confidence interval.It means for any given timepoint, the value of the signal is between ymax and ymin corresponding to x axis(kind of standard deviation-expected deviation)
```

```python
#ignoring confidence interval
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri, ci= False)
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri, ci= 'sd') #showing the standard deviation
```

```python
#plotting without confidence interval
sns.relplot(x='timepoint',y='signal',estimator=None,kind='line',data=fmri)
```

```python
#plot the data with semantic mapping(hue,size,style)
sns.relplot(x='timepoint',y='signal',hue='event',kind='line',data=fmri) #adding categorical value
```

```python
sns.relplot(x='timepoint',y='signal',hue='region',style='event',kind='line',data=fmri)
```

```python
#add markers for easier interpretation
sns.relplot(x='timepoint',y='signal',hue='region',markers=True,style='event',kind='line',data=fmri)
```

```python
sns.relplot(x='timepoint',y='signal',hue='region',markers=True,style='event',kind='line',data=fmri, dashes=False)
```

```python
#adding colors,since here we have added event to both  hue and style, we have difference of colors and style(check how one is solid line and other is dotted )
sns.relplot(x='timepoint',y='signal',hue='event',kind='line',markers=True,style='event',data=fmri)
```
